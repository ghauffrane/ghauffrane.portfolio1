## Feature Engineering

```python
newdt.dtypes
```

```python
Delay_Fix = []
for i in range(0,len(newdt)):
    delay = newdt.iloc[i,5]
    delay = str(delay)
    if(delay == "On Time"):
        delay = delay.replace("On Time", "0")
    else:
        delay = delay.replace("min","")
    Delay_Fix.append(delay)

newdt["status"] = Delay_Fix
newdt["status"] = pd.to_numeric(newdt["status"], errors="coerce")

    
```
